```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv("IPL.csv")

features = ['Venue','team1','team2','toss_winner', 'toss_decision']

df = df[df['winner'].notna()]
df.fillna ('Unknown', inplace=True)

x = df[features]
y = df['winner']

encoder = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), features)
    ]
)

x_encoded = encoder.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(
    x_encoded, y, test_size=0.2, random_state=42
)

model = RandomForestClassifier()
model.fit(x_train, y_train)

prediction = model.predict(x_test)


print("----- IPL Match Prediction-----")
try:
    teams = df['team1'].unique()
    Venue = df['Venue'].unique()
    toss = df['toss_decision'].unique()


    print("\nAvailable Teams:\n")

    for i, team in enumerate(teams):
        print(i, team)

    print("\n Available Venues:\n") 

    for i, venue in enumerate(Venue):
        print(i, venue)

    print("\n Toss Decision:\n")     

    for i, Toss in enumerate(toss):
        print(i, Toss)       

    team1 = teams[int(input("Select team1 number: "))]
    team2 = teams[int(input("Select team2 number: "))]
    Venue = Venue[int(input("Select Venue number: "))]
    toss_winner = teams[int(input("Select Toss Winning Team number: "))]
    toss_decision = toss[int(input("Select Toss decision (bat || field):"))]

    user_input_df = pd.DataFrame([{
        'Venue' : Venue,
        'team1' : team1,
        'team2'  : team2,
        'toss_winner' : toss_winner,
        'toss_decision' : toss_decision


    }])

    user_input_encoded = encoder.transform(user_input_df)

    prediction = model.predict(user_input_encoded)[0]

    print(f"\n So As Per My Model Predicted WINNER  : {prediction}")

except Exception as e:
    print("An error occured", e)      

```
